# Preprocessing Totalseq A fastq files and QC results

```python
from cite_count import AB_count
```

Input directories where files are located; for totalseq A reference please use Totalseq_A_list.csv; you can get the antibody tag sequences from sequence team   

```python
seq_f = './FASTQ/Totalseq_A_v3_R1_001.fastq'
seq_r = './FASTQ/Totalseq_A_v3_R2_001.fastq'
tag_list = 'Totalseq_A_list.csv'
AB_list = ['TATCCCTTGGGATGG', 'TGTTCCCGCTCAACT', 'CTCCGAATCATGTTG','CAATCAGACCTATGA', 'GCTGCGCTTTCCATT', 'GTCCCTGCAACTTGA','TTTGTCCTGTACGCC']
test_AB = AB_count(PROJ_NAME = 'TotalSeq_A_Test')
```

```python
test_AB.process_fastq(seq_f,seq_r,tag_list,AB_list,overwrite = False) # if True the cache file will be overwrittern; if false it will try to read cache file 
```

```python
test_AB.qc_tag_count(dpi = 120,THRESHOLD_UMI = None, saturation='half') # default settings, see "Fastq import and tag counts QC" for details
```

```python
test_AB.filter_tag_count(dpi = 100,P_OUTLIER = 1e-6,view=True)
```

```python
test_AB.get_tag_counts_table(filter_barcodes=True, overwrite=False) # if filter_barcodes=False then filter will not be applied
```

```python
test_AB.qc_tag_label(max_population = 5, min_population_size = 0.2, confidence_interval = 0.95,dpi = 60,verbose=False,view = True) # See function for parameter details
test_AB.get_tag_label()
```

```python
test_AB.df_merge_UMI_mx # The filtered antibody tag counts table
```

```python
test_AB.df_merge_UMI_mx.to_csv('CITE_Seq_Result.csv')
```
